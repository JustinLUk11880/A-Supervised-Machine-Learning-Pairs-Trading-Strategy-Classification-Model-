```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Download data for KO and PEP
tickets = ['KO', 'PEP']
start_date = '2020-01-01'
end_date = '2026-01-01'

price_data = yf.download(tickets, 
                        start=start_date, 
                        end=end_date, 
                        interval="1d",
                        auto_adjust=True)

price_data.head()
```

```python
# Save raw data (closing prices only)
price_close = price_data['Close']
price_close.head()
```

```python
prices = price_close.dropna()
prices.to_csv("../data_raw/ko_pep_prices.csv")
prices.head()
```
