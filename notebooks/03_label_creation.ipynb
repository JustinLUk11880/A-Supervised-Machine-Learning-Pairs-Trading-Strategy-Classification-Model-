```python
import sys
sys.path.append("..")

import pandas as pd
```

```python
# Load features
features = pd.read_csv(
    "../data_processed/features.csv",
    index_col=0,
    parse_dates=True
)

features.head()
```

```python
# predict will mean-revert over the next N days, 1 = yes, 0 = no
N = 5  # prediction horizon in days

z = features["z_score"]
future_abs_z = z.abs().shift(-N)

y = (future_abs_z < z.abs()).astype(int)
y.name = "y"
```

```python
# join the labels
dataset = features.join(y).dropna()
dataset.head()

# Sanity checks
dataset["y"].value_counts(normalize=True)
dataset[["z_score", "y"]].tail(20)

```

```python
# save the dataset
dataset.to_csv("../data_processed/dataset.csv")
print("Saved dataset to data_processed/dataset.csv")
print("Shape:", dataset.shape)
```
