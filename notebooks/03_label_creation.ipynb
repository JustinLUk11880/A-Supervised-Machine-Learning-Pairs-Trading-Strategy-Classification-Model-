```python
import sys
sys.path.insert(0, '../')

from src.labeling import create_labels
import pandas as pd
```

```python
# Load features
ticker1 = 'KO'
ticker2 = 'PEP'

features = pd.read_csv(f'../data_processed/features_{ticker1}_{ticker2}.csv', index_col=0, parse_dates=True)
print(f"Features shape: {features.shape}")
print(features.head())
```

```python
# Create labels based on z-score
spread = features['spread']
zscore = features['zscore']

labels = create_labels(spread, zscore, long_threshold=-1.0, short_threshold=1.0)

print(f"Labels distribution:")
print(labels.value_counts())
print(f"\nLabel percentages:")
print(labels.value_counts(normalize=True) * 100)
```

```python
# Add labels to features
features['label'] = labels

# Save
features.to_csv(f'../data_processed/features_with_labels_{ticker1}_{ticker2}.csv')
print(f"Saved features with labels to data_processed/")
```
